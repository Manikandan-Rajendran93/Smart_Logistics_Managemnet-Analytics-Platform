```python
df3["route_id"] = df3["route_id"].str.strip()
```

```python
df3["destination"] = df3["destination"].str.strip()
```

```python
conn = mysql.connector.connect(host = "localhost", user = "root", password = "9095", database= "smart_logistics")
cursor = conn.cursor()
```

```python
for index, row in df3.iterrows():
    sql = "insert into routes (route_id, origin, destination, distance_km, avg_time_hours) values (%s, %s, %s, %s, %s)"
    cursor.execute(sql, (row["route_id"], row["origin"], row["destination"], row["distance_km"], row["avg_time_hours"]))
conn.commit()
cursor.close()
conn.close()
```
